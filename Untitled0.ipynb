```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Ingresos')
plt.title('Boxplot de Ingresos')
plt.xlabel('Ingresos')
plt.show()
```

En este boxplot que se realizo podemos interpretar que la distribución de los ingresos se encuentra sesgada hacia la derecha lo que en este caso nos podria indicar que hay mas personas con ingresos bajos a medios y que existe una menor cantidad de personas con ingresos altos esto se puede confirmar con la presencia de los valores atipicos que se pueden observar

**Histograma para ingresos**

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Ingresos'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()
```

En este histograma que presenta asimetria positiva podemos ver que la distribucion no es uniforme ya que la mayor parte de los datos se encuentran en los ingresos medios y bajos y que los valores atipicos se encuentran muy por encima de la media

```python
# Número total de campañas aceptadas
df['Total_Accepted'] = df[['Acepto_Campaña_1', 'Acepto_Campaña_2', 'Acepto_Campaña_3', 'Acepto_Campaña_4', 'Acepto_Campaña_5']].sum(axis=1)

# Gráfico de barras de las respuestas a las campañas
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Total_Accepted', palette='viridis')
plt.title('Respuestas a las Campañas')
plt.xlabel('Número de Campañas Aceptadas')
plt.ylabel('Frecuencia')
plt.show()
```

En este grafico podemos ver que las campañas de marketing no han sido buenas ya que se puede ver que la mayoria no ha aceptado las campañas y un muy poco porcentaje lo ha hecho

```python
# Contar el número de hijos (suma de Kidhome y Teenhome)
# Verificar nombres de columnas en el DataFrame


# Ajustar los nombres de las columnas si es necesario
df['Total_Kids'] = df['Hijos_Pequeños'] + df['Adolescentes']

# Gráfico de barras del número total de hijos
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Total_Kids', palette='pastel')
plt.title('Distribución del Número de Hijos')
plt.xlabel('Número de Hijos')
plt.ylabel('Frecuencia')
plt.show()
```

En este grafico podemos ver que la mayoria cuenta con 1 hijo seguido de 0 hijos luego de 2 hijos y como ultimo podemos observar que un porcentaje muy bajo cuenta con 3 hijos

```python
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

# 2. Create a DataFrame with only numerical features:
df_stats = df[numerical_cols]

# 3. Calculate the correlation matrix:
correlation_matrix = df_stats.corr()

# 4. Visualize the correlation matrix using a heatmap:
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, cmap="BrBG", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación de Variables Numéricas")
plt.show()  # This line is essential to display the plot
```
